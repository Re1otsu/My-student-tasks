```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing

# 1. Загружаем данные
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
df = df.drop(columns=['anaemia','diabetes','high_blood_pressure','sex','smoking','time','DEATH_EVENT'])
data = df.to_numpy(dtype='float')

# 2. Настройка стандартизации на первых 150 наблюдениях
scaler_150 = preprocessing.StandardScaler().fit(data[:150, :])

# 3. Стандартизация всех данных
data_scaled_150 = scaler_150.transform(data)

# 4. Гистограммы после стандартизации
n_bins = 20
fig, axs = plt.subplots(2, 3, figsize=(12, 8))

axs[0, 0].hist(data_scaled_150[:, 0], bins=n_bins); axs[0, 0].set_title('age')
axs[0, 1].hist(data_scaled_150[:, 1], bins=n_bins); axs[0, 1].set_title('creatinine_phosphokinase')
axs[0, 2].hist(data_scaled_150[:, 2], bins=n_bins); axs[0, 2].set_title('ejection_fraction')
axs[1, 0].hist(data_scaled_150[:, 3], bins=n_bins); axs[1, 0].set_title('platelets')
axs[1, 1].hist(data_scaled_150[:, 4], bins=n_bins); axs[1, 1].set_title('serum_creatinine')
axs[1, 2].hist(data_scaled_150[:, 5], bins=n_bins); axs[1, 2].set_title('serum_sodium')

plt.tight_layout()
plt.show()

# 5. Мат. ожидание и СКО до и после стандартизации
mean_before = np.mean(data, axis=0)
std_before = np.std(data, axis=0)

mean_after = np.mean(data_scaled_150, axis=0)
std_after = np.std(data_scaled_150, axis=0)

results = pd.DataFrame({
    "Признак": df.columns,
    "Среднее до": mean_before,
    "СКО до": std_before,
    "Среднее после": mean_after,
    "СКО после": std_after
})
print(results.round(3))

# 6. Формулы стандартизации + сравнение с scaler
print("\nФормулы стандартизации (по 150 наблюдениям):")
for col, mu, sigma in zip(df.columns, scaler_150.mean_, np.sqrt(scaler_150.var_)):
    print(f"{col}: z = (x - {mu:.2f}) / {sigma:.2f}")

print("\nЗначения из scaler:")
for col, mu, var in zip(df.columns, scaler_150.mean_, scaler_150.var_):
    print(f"{col}: mean_ = {mu:.2f}, sigma = {np.sqrt(var):.2f}")

# 7. Настройка стандартизации на всех данных
scaler_all = preprocessing.StandardScaler().fit(data)
data_scaled_all = scaler_all.transform(data)

compare = pd.DataFrame({
    "Признак": df.columns,
    "mean (150)": scaler_150.mean_,
    "sigma (150)": np.sqrt(scaler_150.var_),
    "mean (all)": scaler_all.mean_,
    "sigma (all)": np.sqrt(scaler_all.var_)
})
print("\nСравнение параметров (150 наблюдений vs все данные):")
print(compare.round(3))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing

# Загружаем данные
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
df = df.drop(columns=['anaemia','diabetes','high_blood_pressure','sex','smoking','time','DEATH_EVENT'])
data = df.to_numpy(dtype='float')

# 1. MinMaxScaler
min_max_scaler = preprocessing.MinMaxScaler().fit(data)
data_min_max_scaled = min_max_scaler.transform(data)

# 2. Гистограммы (MinMaxScaler vs исходные данные)
n_bins = 20
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
for i, col in enumerate(df.columns):
    r, c = divmod(i, 3)
    axs[r, c].hist(data_min_max_scaled[:, i], bins=n_bins, alpha=0.7, label="MinMax")
    axs[r, c].hist(data[:, i], bins=n_bins, alpha=0.5, label="Original")
    axs[r, c].set_title(col)
    axs[r, c].legend()
plt.tight_layout()
plt.show()

# 3. Минимум и максимум из параметров MinMaxScaler
print("Min и Max для каждого признака (MinMaxScaler):")
for col, data_min, data_max in zip(df.columns, min_max_scaler.data_min_, min_max_scaler.data_max_):
    print(f"{col}: min = {data_min:.2f}, max = {data_max:.2f}")

# 4. MaxAbsScaler и RobustScaler
maxabs_scaler = preprocessing.MaxAbsScaler().fit(data)
data_maxabs_scaled = maxabs_scaler.transform(data)

robust_scaler = preprocessing.RobustScaler().fit(data)
data_robust_scaled = robust_scaler.transform(data)

# Гистограммы для MaxAbs и Robust
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
for i, col in enumerate(df.columns):
    r, c = divmod(i, 3)
    axs[r, c].hist(data_maxabs_scaled[:, i], bins=n_bins, alpha=0.7, label="MaxAbs")
    axs[r, c].hist(data[:, i], bins=n_bins, alpha=0.5, label="Original")
    axs[r, c].set_title(col)
    axs[r, c].legend()
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 3, figsize=(12, 8))
for i, col in enumerate(df.columns):
    r, c = divmod(i, 3)
    axs[r, c].hist(data_robust_scaled[:, i], bins=n_bins, alpha=0.7, label="Robust")
    axs[r, c].hist(data[:, i], bins=n_bins, alpha=0.5, label="Original")
    axs[r, c].set_title(col)
    axs[r, c].legend()
plt.tight_layout()
plt.show()

print("\nОсобенности:")
print("- MinMaxScaler приводит данные в диапазон [0, 1].")
print("- MaxAbsScaler приводит данные к диапазону [-1, 1], деля на максимальное по модулю значение.")
print("- RobustScaler масштабирует данные относительно медианы и межквартильного размаха (устойчив к выбросам).")

# 5. Функция для приведения данных к диапазону [-5, 10]
def scale_to_range(data, feature_range=(-5, 10)):
    custom_scaler = preprocessing.MinMaxScaler(feature_range=feature_range)
    return custom_scaler.fit_transform(data)

data_custom_scaled = scale_to_range(data, (-5, 10))

print("\nПример диапазона после приведения к [-5, 10]:")
for i, col in enumerate(df.columns):
    print(f"{col}: min = {data_custom_scaled[:, i].min():.2f}, max = {data_custom_scaled[:, i].max():.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing

# Загружаем данные
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
df = df.drop(columns=['anaemia','diabetes','high_blood_pressure','sex','smoking','time','DEATH_EVENT'])
data = df.to_numpy(dtype='float')

n_bins = 20

# 1. QuantileTransformer (равномерное распределение)
quantile_transformer = preprocessing.QuantileTransformer(n_quantiles=100, random_state=0)
data_quantile_scaled = quantile_transformer.fit_transform(data)

# 2. Гистограммы (равномерное распределение)
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
for i, col in enumerate(df.columns):
    r, c = divmod(i, 3)
    axs[r, c].hist(data[:, i], bins=n_bins, alpha=0.5, label="Original")
    axs[r, c].hist(data_quantile_scaled[:, i], bins=n_bins, alpha=0.7, label="Quantile [uniform]")
    axs[r, c].set_title(col)
    axs[r, c].legend()
plt.tight_layout()
plt.show()

print("n_quantiles = 100 означает, что данные разбиваются на 100 квантилей.")
print("→ Чем больше n_quantiles, тем ближе распределение будет к равномерному.")
print("→ При маленьком n_quantiles распределение будет ступенчатым (дискретным).")

# 4. QuantileTransformer (нормальное распределение)
quantile_normal = preprocessing.QuantileTransformer(n_quantiles=100, 
                                                    random_state=0, 
                                                    output_distribution="normal")
data_quantile_normal = quantile_normal.fit_transform(data)

# 5. Гистограммы (нормальное распределение)
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
for i, col in enumerate(df.columns):
    r, c = divmod(i, 3)
    axs[r, c].hist(data[:, i], bins=n_bins, alpha=0.5, label="Original")
    axs[r, c].hist(data_quantile_normal[:, i], bins=n_bins, alpha=0.7, label="Quantile [normal]")
    axs[r, c].set_title(col)
    axs[r, c].legend()
plt.tight_layout()
plt.show()

# 6. PowerTransformer (приведение к нормальному распределению)
power_transformer = preprocessing.PowerTransformer(method="yeo-johnson", standardize=True)
data_power_scaled = power_transformer.fit_transform(data)

# Гистограммы для PowerTransformer
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
for i, col in enumerate(df.columns):
    r, c = divmod(i, 3)
    axs[r, c].hist(data[:, i], bins=n_bins, alpha=0.5, label="Original")
    axs[r, c].hist(data_power_scaled[:, i], bins=n_bins, alpha=0.7, label="PowerTransformer")
    axs[r, c].set_title(col)
    axs[r, c].legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import KBinsDiscretizer

# Загружаем данные
df = pd.read_csv("heart_failure_clinical_records_dataset.csv")

# Оставим только нужные признаки
features = ["age", "creatinine_phosphokinase", "ejection_fraction", 
            "platelets", "serum_creatinine", "serum_sodium"]

X = df[features]

# Настройка количества интервалов для каждого признака
n_bins = [3, 4, 3, 10, 2, 4]

# Инициализация и обучение дискретизатора
kbd = KBinsDiscretizer(n_bins=n_bins, encode="ordinal", strategy="uniform")
X_binned = kbd.fit_transform(X)

# Переводим обратно в DataFrame для удобства
X_binned_df = pd.DataFrame(X_binned, columns=features)

# --- 1. Построение гистограмм ---
fig, axs = plt.subplots(2, 3, figsize=(15, 8))
axs = axs.ravel()

for i, col in enumerate(features):
    axs[i].hist(X_binned_df[col], bins=n_bins[i], edgecolor="black")
    axs[i].set_title(f"{col} (дискретизировано)")
    axs[i].set_xlabel("bin index")
    axs[i].set_ylabel("count")

plt.tight_layout()
plt.show()

# --- 2. Вывод границ интервалов ---
for feat, edges in zip(features, kbd.bin_edges_):
    print(f"{feat}: {edges}")
```
