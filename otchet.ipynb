```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return x**2 + 2*x - 4

def dichotomy_search(a, b, tol=1e-3, delta=1e-4, max_iter=100):
    iterations = 0
    while (b - a) / 2 > tol and iterations < max_iter:
        mid = (a + b) / 2
        x1 = mid - delta
        x2 = mid + delta
        f1, f2 = f(x1), f(x2)
        if f1 < f2:
            b = x2
        else:
            a = x1
        iterations += 1
    x_min = (a + b) / 2
    return x_min, f(x_min), iterations

a, b = -2, 1
x_min, f_min, iters = dichotomy_search(a, b)

print(f"Функция минимумы: x* = {x_min:.4f}")
print(f"Функция мәні: f(x*) = {f_min:.4f}")
print(f"Итерация саны: {iters}")

x = np.linspace(a, b, 200)
y = f(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='f(x) = x² + 2x - 4', color='blue')
plt.scatter(x_min, f_min, color='red', s=80, label=f'Минимум: x*={x_min:.3f}')
plt.title("Дихотомия әдісі — функция минимумын табу")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 2*x - 4

def fibonacci(n):
    fib = [1, 1]
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib

def fibonacci_search(a, b, tol=1e-3):
    fib = [1, 1]
    while fib[-1] < (b - a) / tol:
        fib.append(fib[-1] + fib[-2])
    n = len(fib)

    x1 = a + (fib[n - 3] / fib[n - 1]) * (b - a)
    x2 = a + (fib[n - 2] / fib[n - 1]) * (b - a)
    f1, f2 = f(x1), f(x2)

    for k in range(1, n - 2):
        if f1 > f2:
            a = x1
            x1 = x2
            f1 = f2
            x2 = a + (fib[n - k - 2] / fib[n - k - 1]) * (b - a)
            f2 = f(x2)
        else:
            b = x2
            x2 = x1
            f2 = f1
            x1 = a + (fib[n - k - 3] / fib[n - k - 1]) * (b - a)
            f1 = f(x1)
        if abs(b - a) < tol:
            break

    x_min = (a + b) / 2
    return x_min, f(x_min), k

a, b = -2, 1

x_min, f_min, iters = fibonacci_search(a, b)

print(f"Функция минимумы: x* = {x_min:.4f}")
print(f"Функция мәні: f(x*) = {f_min:.4f}")
print(f"Итерация саны: {iters}")

x = np.linspace(a, b, 200)
y = f(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='f(x) = x² + 2x - 4', color='blue')
plt.scatter(x_min, f_min, color='red', s=80, label=f'Минимум: x*={x_min:.3f}')
plt.title("Фибоначчи әдісі — функция минимумын табу")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 2*x - 4

def df(x):
    return 2*x + 2

def secant_method(x0, x1, tol=1e-6, max_iter=100):
    for i in range(max_iter):
        f0 = df(x0)
        f1 = df(x1)
        if abs(f1 - f0) < 1e-12:
            break
        x2 = x1 - f1 * (x1 - x0) / (f1 - f0)
        if abs(x2 - x1) < tol:
            return x2, f(x2), i + 1
        x0, x1 = x1, x2
    return x2, f(x2), max_iter

a, b = -2, 1

x_min, f_min, iters = secant_method(a, b)

print(f"Функция минимумы: x* = {x_min:.4f}")
print(f"Функция мәні: f(x*) = {f_min:.4f}")
print(f"Итерация саны: {iters}")

x = np.linspace(a, b, 200)
y = f(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='f(x) = x² + 2x - 4', color='blue')
plt.scatter(x_min, f_min, color='red', s=80, label=f'Минимум: x*={x_min:.3f}')
plt.title("Секанттық әдісі — функция минимумын табу")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
